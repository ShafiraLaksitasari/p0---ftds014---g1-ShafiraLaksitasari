## Working Area

### Data Loading

Import pandas before processing.

```python
import numpy as np
import pandas as pd
```

Import the provided datas from library.

```python
pokemon = pd.read_csv('GC Data/Pokemon.csv')
pokemon
```

### Data Cleaning

Must pay attention to the null data and data type for each column.

```python
pokemon.info()
```

Apparently there are several null data from the information above. It is important to explore which the data column are null and how many of them are null.

```python
pokemon.isnull().sum()
```

Mostly the null are from Type 2 Pokemon because not every Pokemon has 2 elements. Because of this situation, replace the 'Nan' into 'None' to avoid any null data.

```python
pokemon['Type 2'].fillna('None', inplace=True)
pokemon
```

There are no more null data from the data frame if the total of null is schecked again.

```python
pokemon.isnull().sum()
```

Asssuming the Pokemon is included as legendary or not and the generation where its come from are fully ignored.

```python
pokemon.drop(labels=['Generation','Legendary'], axis = 1, inplace=True)
pokemon
```

Make the ID in each of pokemon is different even the Pokemon name is similiar to make it unique.

```python
pokemon['#'] = np.arange(len(pokemon)) + 1
pokemon
```

```python
pokemon.rename(columns = {'#': 'ID'}, inplace=True)
pokemon
```

Change from the default index to '#'

```python
pokemon = pokemon.set_index('ID')
```

```python
pokemon
```

Make sure the total is based on total value of HP, Attack, Defense, Special Attack, Special Defense, and Speed

```python
pokemon['Total'] = pokemon['HP'] + pokemon['Attack'] + pokemon['Defense'] + pokemon['Sp. Atk'] + pokemon['Sp. Def'] + pokemon['Speed']
pokemon
```

### Data Exploration

```python
pokemon
```

##### Finding the top tenth Pokemon based on Total value. 

Atleast it is better to obtain the Pokemon which total value more than the total average value. Find the mean first.

```python
pokemon['Total'].mean()
```

Then all the pokemon which Total more than average can be obtained.

```python
pok_total = pokemon[pokemon["Total"] > 435.1025]
pok_total
```

Sorting the data frame 'pok_total' above with descending to obtain the Top tenth of powerful pokemon based on the total.

```python
pok_total.sort_values(
     by="Total",
     ascending=False
).head(10)
```

MewtwoMega Mewtwo, MewtwoMega Mewtwo X, and RayquazaMega Rayquaza founded as the strongest Pokemon based on its maximun Total. Some pokemon are superior only in one or a few abilities. For a little balance, atleast finding all the Pokemon atleast more than average value in each ability

#### Finding the top tenth Pokemon based on mean

Find the average in each ability.

```python
print(pokemon['HP'].mean())
print(pokemon['Attack'].mean())
print(pokemon['Defense'].mean())
print(pokemon['Sp. Atk'].mean())
print(pokemon['Sp. Def'].mean())
print(pokemon['Speed'].mean())
```

Obtain all the pokemon wich fulfilled the condition.

```python
# Query
pok_mean = pokemon[(pokemon['HP'] > 69.25875) &
(pokemon['Attack'] > 79.00125) &
(pokemon['Defense'] > 73.8425) &
(pokemon['Sp. Atk'] > 72.82) &
(pokemon['Sp. Def'] > 71.9025) &
(pokemon['Speed'] > 68.2775)
]
pok_mean
```

Sorting the result of table 'pok_mean' above based on the highest Total.

```python
pok_mean.sort_values(
     by="Total",
     ascending=False
).head(15)
```

The top tenth Pokemon based on each mean are obtained.

#### Finding the Strongest Pokemon in 1 Type and 2 Type

Obtain the Pokemon which only have 1 Type.

```python
pok_1type = pokemon[(pokemon['Type 2'] == 'None') &
(pokemon['HP'] > 69.25875) &
(pokemon['Attack'] > 79.00125) &
(pokemon['Defense'] > 73.8425) &
(pokemon['Sp. Atk'] > 72.82) &
(pokemon['Sp. Def'] > 71.9025) &
(pokemon['Speed'] > 68.2775)
]
pok_1type
```

To Obtain the top tenth 1 Type strongest Pokemon.

```python
pok_1type.sort_values(
     by="Total",
     ascending=False
).head(10)
```

Obtain the Pokemon which have 2 Type.

```python
pok_2type = pokemon[(pokemon['Type 2'] != 'None') &
(pokemon['HP'] > 69.25875) &
(pokemon['Attack'] > 79.00125) &
(pokemon['Defense'] > 73.8425) &
(pokemon['Sp. Atk'] > 72.82) &
(pokemon['Sp. Def'] > 71.9025) &
(pokemon['Speed'] > 68.2775)
]
pok_2type
```

To Obtain the top tenth 2 Type strongest Pokemon.

```python
pok_2type.sort_values(
     by="Total",
     ascending=False
).head(10)
```

#### Finding the top element or Type

There are several types in pokemon

```python
pokemon[["Type 1","Type 2"]].value_counts()
```

The pokemon elemen can become weakness or strength for other elemen type. For example the opponents has water elements. Just like scissor, rock, paper game, water weak towards ice and grass.  If we want to know whch Pokemon is the strongest based on Ice or Grass Type, we can see from diagram below.

First, to obtain the pkemon which has element that must be on our team we can find the most owned type.

```python
tipe_1 = pokemon["Type 1"].value_counts()
tipe_2 = pokemon["Type 2"].value_counts()
```

```python
tipe_1_column = pd.DataFrame(tipe_1)
#tipe_1_column.rename(columns = {'Type 1': 'Type'}, inplace=True)
tipe_1_column
```

```python
tipe_2_column = pd.DataFrame(tipe_2)
#tipe_2_column.rename(columns = {'Type 2': 'Type'}, inplace=True)
tipe_2_column
```

Merge the type between type 1 and type 2

```python
all_pokemon_type = pd.concat([tipe_1_column, tipe_2_column], axis=1, sort=False)
all_pokemon_type
```

Then obtained the total of pokemon for each element (with null value change into 0)

```python
all_pokemon_type['Type 1'].fillna(0, inplace=True)

total_pokemon = all_pokemon_type["Type 1"] + all_pokemon_type["Type 2"]

all_pokemon_type["Total of Pokemon"] = total_pokemon
all_pokemon_type
```

After calculate the total, we got the most type owned by pokemon after the sorting from the data frame below.

```python
sorted_all_pokemon_type = all_pokemon_type.sort_values(
     by="Total of Pokemon",
     ascending=False
)
sorted_all_pokemon_type
```

None following by water and normal is the most type of pokemon. Since none is not type and normal can be attacked by anything, it also can be seen water is the most common element. Based on infromation from one of Pokemon "Expert", water weakness is Ice and Grass. So we can obtain Pokemon which has these type to include in our team.

```python
group_Type1 = pokemon.groupby(by='Type 1')
group_Type2 = pokemon.groupby(by='Type 2')
```

```python
grass_tipe1 = group_Type1.get_group('Grass')
ice_tipe1 = group_Type1.get_group('Ice')
grass_tipe2 = group_Type2.get_group('Grass')
ice_tipe2 = group_Type2.get_group('Ice')
```

```python
water_weakness = pd.concat([grass_tipe1, ice_tipe1, grass_tipe2, ice_tipe2], axis=0)
water_weakness
```
